## Plotting Sinusoidal

```

import numpy as np
import matplotlib.pyplot as plt
import math as m

t = np.arange(-1*m.pi,m.pi,0.01);
print(t.shape)
print(t[0:15])
```

### Plotting Sin with period of 2*pi and amplitude of 1 

```
a   = np.sin(t)
plt.plot(t,a)
plt.xlabel('Angle in rad')
plt.ylabel('Amplitude')
plt.grid(True, which='both')
plt.axis('tight')
plt.show()
```

```
a   = 4*np.sin(t)
plt.plot(t,a)
plt.xlabel('Angle in rad')
plt.ylabel('Amplitude')
plt.grid(True, which='both')
plt.axis('tight')
plt.show()
```

```
a   = 4*np.sin(2*t)
plt.plot(t,a)
plt.xlabel('Angle in rad')
plt.ylabel('Amplitude')
plt.grid(True, which='both')
plt.axis('tight')
plt.show()
```
